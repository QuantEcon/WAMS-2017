### Exercise

Create a function `AR1(rho, T)` that simulates an AR1 model with parameter $\alpha$ and returns an array of length `n`

$$
x_{t} = \rho x_{t-1} + \varepsilon_t \quad where \quad x_0 = 0
$$

```julia
function AR1(rho, T)
    x = zeros(T + 1)
    for t in 1:T
        x[t+1] = rho * x[t] + randn()
    end
    return x
end
```

```julia
AR1(0.9, 10)
```

```julia
using Plots

plot(AR1(0.9, 100))
```
